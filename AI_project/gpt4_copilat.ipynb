```python
from datetime import datetime

# Initialize attendance dictionary for one month
attendance = {day: "Absent" for day in range(1, 31)}

def mark_attendance(day, status):
    if 1 <= day <= 30:
        attendance[day] = status
        print(f"Attendance for day {day} marked as {status}.")
    else:
        print("Invalid day. Please enter a day between 1 and 30.")

def display_attendance():
    print("\nAttendance for the month:")
    for day, status in attendance.items():
        print(f"Day {day}: {status}")

# Example usage
while True:
    print("\nAttendance System")
    print("1. Mark Attendance")
    print("2. Display Attendance")
    print("3. Exit")
    choice = input("Enter your choice: ")

    if choice == "1":
        try:
            day = int(input("Enter the day (1-30): "))
            status = input("Enter status (Present/Absent): ")
            mark_attendance(day, status)
        except ValueError:
            print("Invalid input. Please enter a valid day and status.")
    elif choice == "2":
        display_attendance()
    elif choice == "3":
        print("Exiting the attendance system.")
        break
    else:
        print("Invalid choice. Please try again.")
```
